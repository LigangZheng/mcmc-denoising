```python
import denoising
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread
```

```python
lars = imread("data/lars-noisy-gray.jpg")[:, :, 0]/255
```

```python
plt.imshow(lars)
```

```python
pottsobj = denoising.PottsMH(lars, 4, lars.std(), 10)
```

```python
result = np.mean([pottsobj.solve(20) for i in range(20)], 0)
```

```python
plt.figure(figsize=(20, 20))

plt.imshow(np.array(result), cmap="gray")
```
